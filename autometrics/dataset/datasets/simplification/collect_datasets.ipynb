```python
import pandas as pd
```

```python
def add_ref_columns(df, ref_dir, input_col='text'):
    # References are organized into two files: test.dst and test.src
    # src contains the original text, dst contains the reference simplifications separated by tabs

    # Load the reference files
    src_file = ref_dir + '/test.src'
    dst_file = ref_dir + '/test.dst'
    src = open(src_file, 'r').read().splitlines()
    dst = open(dst_file, 'r').read().splitlines()

    # Create a dictionary mapping the original text to the reference simplifications
    max_num_refs = 0
    ref_dict = {}
    for i in range(len(src)):
        src_text = src[i].strip()
        ref_dict[src_text] = dst[i].split('\t')
        num_refs = len(dst[i].split('\t'))
        if num_refs > max_num_refs:
            max_num_refs = num_refs

    # Add the reference simplifications to the dataframe
    for i, row in df.iterrows():
        text = row[input_col]
        text = text.strip()
        if text in ref_dict:
            refs = ref_dict[text]
            for j in range(max_num_refs):
                if j < len(refs):
                    df.at[i, 'ref' + str(j + 1)] = refs[j]
                else:
                    df.at[i, 'ref' + str(j + 1)] = ''

    return df
```

```python
# Simpda
simpda_dir = './simpda'
simpda_file = simpda_dir + '/simpDA_2022.csv'
simpda_df = pd.read_csv(simpda_file)
simpda_df = add_ref_columns(simpda_df, simpda_dir + '/references', input_col='Input.original')
simpda_df.to_csv("simpda.csv", index=False)
```

```python
# SimpEval
simpeval_dir = './simpeval'
simpeval_file = simpeval_dir + '/simpeval_2022.csv'
simpeval_df = pd.read_csv(simpeval_file)
simpeval_df = add_ref_columns(simpeval_df, simpeval_dir + '/references', input_col='original')
simpeval_df.to_csv("simpeval.csv", index=False)
```
