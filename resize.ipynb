```python
from autometrics.dataset.datasets.cogym.cogym import CoGymTravelOutcome
from autometrics.dataset.datasets.evalgen.evalgen import EvalGenProduct
from platformdirs import user_data_dir
import pandas as pd
import os
```

```python
dataset = EvalGenProduct()
target_column = "grade" #"grade" #"outcomeRating"
```

```python
train, dev, test = dataset.load_permanent_splits()
```

```python
os.listdir(user_data_dir("autometrics") + '/datasets/' + dataset.name)
```

```python
train.get_dataframe().dropna(subset=[target_column])
```

```python
dev.get_dataframe().dropna(subset=[target_column])
```

```python
print(len(train.get_dataframe().dropna(subset=[target_column])))
print(len(dev.get_dataframe().dropna(subset=[target_column])))
print(len(test.get_dataframe().dropna(subset=[target_column])))
```

```python
difference = len(train.get_dataframe().dropna(subset=[target_column])) - len(dev.get_dataframe().dropna(subset=[target_column]))
```

```python
dropna_train = train.get_dataframe().dropna(subset=[target_column])
dropna_dev = dev.get_dataframe().dropna(subset=[target_column])
```

```python
move_subset = dropna_train.sample(n=(difference // 2))
dropna_train = dropna_train.drop(move_subset.index)
dropna_dev = pd.concat([dropna_dev, move_subset])

print(len(dropna_train))
print(len(dropna_dev))
```

```python
dropna_train.to_csv(user_data_dir("autometrics") + '/datasets/' + dataset.name + '/train_resized.csv', index=False)
dropna_dev.to_csv(user_data_dir("autometrics") + '/datasets/' + dataset.name + '/dev_resized.csv', index=False)
```
